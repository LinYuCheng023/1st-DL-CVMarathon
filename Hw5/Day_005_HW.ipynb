```python
with open(r'imagenet_urls_examples.txt', encoding="utf-8") as f:
    data = f.readlines()
```

```python
from skimage import io

No=[]
content=[]
array_of_img = []
for d in data:
    tmp = d.split('\t')
    No.append(str(tmp[0]))
    tmp2 = tmp[1].split('\n')
    content.append(str(tmp2[0]))
my_dict = {'No.':No,'data':content}
my_df2 = pd.DataFrame(my_dict)

five_data = my_df2[:5]

for i in five_data['data']:
    
    print(i)
    img = io.imread(i)
    io.imshow(img)
    io.show()

```
