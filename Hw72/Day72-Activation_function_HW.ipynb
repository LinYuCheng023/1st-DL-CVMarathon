# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    return max(0,x)

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
x = plt.linspace(-10,10,100)

py = [ReLU(xv) for xv in x]

plt.plot(x,py,'b',label='linspace(-10,10,10)')


plt.grid()

plt.title('ReLU Function')

```

```python
def dReLU(x):
    if x>0:
        return 1
    else:
        return 0
        
py2 = [dReLU(xv) for xv in x]

plt.plot(x,py2,'b',label='linspace(-10,10,10)')


plt.grid()

plt.title('dReLU Function')
```
